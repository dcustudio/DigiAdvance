```python
# First, lets import matplotlib for data visualisation and pandas for reading the data.
import matplotlib.pyplot as plt
import pandas as pd
# Now, lets read the data and extract the fields we would like to draw
data = pd.read_csv("IE_Import_Export.csv")

# month = data["Month"]
# temprature = data["Temprature"]

#  an alternative option to pandas data series is Numpy arrays. we need to convert
#  the panda series to a numpy array using values as follows. 

month = data.loc[ : , "Month"].values
imports = data.loc[ : , "Import"].values
exports = data.loc[ : , "Export"].values
```

```python
# We can also check the type of the data values.
print(type(month))
print(type(imports))

# Now both are numpy arrays
```

```python
# Lets plot a simple plot of the data where the x-axix is the month and the y-axis is the temprature data.
plt.plot(month, imports)
```

```python
# Lets plot a simple plot of the data where the x-axix is the month and the y-axis is the imports data.
plt.plot(month, imports, marker = "*")
```

```python
# Lets change the color of the line.
plt.plot(month, imports, c = "r", marker = "*")
```

```python
# Customisation: adding title and x and y labels

# Use xlabel to add x-axis label
plt.xlabel("Month")

# Use ylabel to add the y-axis label
plt.ylabel("in Thousand Euros")

# Use title to add a title to your graph
plt.title("Monthly Import Amount, Ireland 2021")
plt.plot(month, imports, c = "r", marker = "*")
```

```python
# Customisation: increasing the font size of title and its color to blue

plt.xlabel("Month")
plt.ylabel("in Thousand Euros")

# Use the size parameter to indicate the font size
plt.title("Monthly Import Amount, Ireland 2021", size = 20, c = "b")
plt.plot(month, imports, c = "r", marker = "*")
```

```python
# Now lets add limits to the x and y axis to change the lower/upper or left/right numbers that show up on the plot
plt.xlabel("Month")
plt.ylabel("in Thousand Euros")
plt.title("Monthly Import Amount, Ireland 2021", size = 20, c = "b")
plt.xlim(0, 12)
plt.plot(month, imports, c = "r", marker = "+")
```

```python
# Look at the x-axis and y-axis ticks, the y-axis is written in scientific notation and the x axis is a bit cluttered.

plt.xlabel("Month")
plt.ylabel("In Thousand Euros")
plt.title("Monthly Import Amount, Ireland 2021", size = 20, c = "b")


# plot the grpah
plt.plot(month, imports, c = "r", marker = "*")

# the xlim determines min and max values that will show up on the x-axis ticks.
plt.xlim(0, 11)

# rotate the x-axis label to remov ethe clutter
plt.xticks(month, rotation = 90)  # you can also specify rotation ="vertical"

# turn off the scientific notation to format the numbers and add a comma as a separator
y_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(["{:,.0f}".format(x) for x in y_values])
```

```python
# Let add grid lines to make the data points further readable

plt.xlabel("Month")
plt.ylabel("In Thousand Euros")
plt.title("Monthly Import Amount, Ireland 2021", size = 20, c = "b")


# plot the grpah
plt.plot(month, imports, c = "r", marker = "*")

# the xlim determines min and max values that will show up on the x-axis ticks.
plt.xlim(0, 11)

# rotate the x-axis label to remov ethe clutter
plt.xticks(month, rotation = 90)  # you can also specify rotation ="vertical"

# turn off the scientific notation to format the numbers
y_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(["{:,.0f}".format(x) for x in y_values])

plt.grid()

```

```python
# Sometimes, we need to show the data values on the lines to improve the readability.

plt.xlabel("Month")
plt.ylabel("In Thousand Euros")
plt.title("Monthly Import Amount, Ireland 2021", size = 20, c = "b")

plt.plot(month, imports, c = "r", marker = "*")
plt.xlim(0, 11)
plt.xticks(month, rotation = 90)  # you can also specify rotation ="vertical"
y_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(["{:,.0f}".format(x) for x in y_values])

plt.grid()

for i, j in zip(month, imports):
   plt.text(i, j, "{}".format( j))
```

```python
# Sometimes, we need to hide the vertical or horizontal grid, to do so use the axis parameter with grid() 

plt.xlabel("Month")
plt.ylabel("In Thousand Euros")
plt.title("Monthly Import Amount, Ireland 2021", size = 20, c = "b")


plt.plot(month, imports, c = "r", marker = "*")
plt.xlim(0, 11)
plt.xticks(month, rotation = 90)  # you can also specify rotation ="vertical"

y_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(["{:,.0f}".format(x) for x in y_values])

# only shows the horizontal grid lines
plt.grid(axis = "y")

# only shows the vertical grid lines
# plt.grid(axis = "x")

for i, j in zip(month, imports):
   plt.text(i, j, "{:,.0f}".format( j))
```
