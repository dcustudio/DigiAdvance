```python
#  Here we demonstrate how we can identify an outlier value and what todo with it.
```

```python
import pandas as pd

# read the data from a file using pandas
data = pd.read_csv("DLAT-Sample-12-2-2.csv")

# The following code checks whether the transport column contains null value or not
data.head(15)
```

```python
# Let's check if a given value is greater than a the inter quartile range (Q3 -Q1)
# compute the first and the third quartile using the following formula

q1 = data['Duration'].quantile(0.25)
q3 = data['Duration'].quantile(0.75)

# Compute the interquartile range

irq = q3 - q1
ol = irq * 1.5

print("first and third quartiles and outlier range,", q1, q3, ol)

#  Now, filter any row containing duration less than q1 -ol and anything greater than q3 +ol

outliers = data[((data['Duration'] < (q1 - ol)) | (data['Duration'] > (q3 + ol)) )]
outliers.head()
```

```python
# using the drop method, you can easily drop all the rows that contain the indexes in the outlier data outliers.
# your final dataset will remove all the rows that contain outlier duration value.

data.drop(outliers.index, inplace=True)
data.head(15)
```
