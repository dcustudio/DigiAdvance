```python
# First, lets import matplotlib for data visualisation and pandas for reading the data.
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Now, let's read the data and extract the fields we would like to draw
data = pd.read_csv("IE_Import_Export.csv")

month = data.loc[ : , "Month"].values
imports = data.loc[ : , "Import"].values
exports = data.loc[ : , "Export"].values


plt.xlabel("Month")
plt.ylabel("In Thousand Euros")
plt.title('Monthly Import Amount, Ireland 2021', size = 20, c = 'b')

bar_width = 0.4

# lets add the name of the months as a label
month_names = ['Jan', 'Feb','Mar','Apr','May', 'Jun','Jul','Aug','Sep', 'Oct','Nov','Dec']
X_axis = np.arange(len(month))

# Set the position of the x_axis labels and the rotation
plt.xticks(X_axis, month_names, rotation = 45)

# You can set the color, width, edgecolor and label of the bar as follows

# plot half way to the left and half way to the right
plt.bar(X_axis - 0.2, imports, color ='#55aaff', width = bar_width, label = 'Imports')
plt.bar(X_axis + 0.2, exports, color ='#EF9A9A', width = bar_width, label = 'Exports')
y_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in y_values])
# automatically set the best location for the legend
plt.legend(loc = 0);

plt.savefig("Monthly_import.png", dpi = 300, bbox_inches = 'tight')
```
